# OpenCV
동영상(이미지)를 추출하거나 여러가지 옵션을 주어 데이터를 찾아내는 오픈소스 라이브러리입니다.

```python
import urllib.request as req
import matplotlib.pyplot as plt
url = "https://t1.daumcdn.net/daumtop_chanel/op/20170315064553027.png"
req.urlretrieve(url, "test.png")


# pip install opencv-python
# OpenCV로 읽어 들이기
import cv2
img = cv2.imread("test.png")
print(img)

plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.show()
cv2.imwrite("out.png", img)

# 이미지 자르기
im2 =img[50:250, 100:300]
cv2.imwrite("cut-resize.png", im2)

# 색공간을 그레이스케일로 변환하기
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

#이미지출력하기
plt.imshow(img, cmap= "gray")
plt.axis("off")  # 옆에 좌표 없애는 함수 
plt.show()


# 네거티브 반전
img = 255 - img # 255가 흰색   흰색 - 이미지색 
#이미지 출력하기
plt.imshow(img, cmap= "gray")
plt.show()

#역치 지정하기
th = 90
img[img > th] = 255
img[img < th] = 0
# 이미지 출력하기
plt.imshow(img, cmap="gray")
plt.show()

# 이미지 크기 변경하기 
im2 = cv2.resize(img,(600,300))
#이미지 출력하기
plt.imshow(cv2.cvtColor(im2, cv2.COLOR_BGR2RGB))
plt.show()
```
